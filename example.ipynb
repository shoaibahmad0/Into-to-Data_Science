```python
#cleaning the missing value

import pandas as pd

df=pd.read_csv('datasets/presidents.csv')
df.head(10)
```

```python
#here we can create a column of the prsident column

df["first"]=df["President"]

#now replace the white spaces
df["first"]=df["first"].replace("[ ].*", "", regex=True)
df.head()
```

```python
import pandas as pd

# Define the function to split the name
def splitname(row):
    # Split the name into first and last name
    row['First'] = row['President'].split(" ")[0]
    row['Last'] = row['President'].split(" ")[-1]
    # Return the modified row
    return row

# Apply the function to each row of the DataFrame
df = df.apply(splitname, axis='columns')

# Delete the 'first' column created earlier
del df['first']
    
df=df.apply(splitname,axis='columns')
df.head()

    
```

```python
# now del the first and last columns

del(df['First'])
del(df['Last'])

pattern="(^[\w]*)(?:.* )([\w]*$)"

df["President"].str.extract(pattern).head()
```

```python
# So that looks pretty nice, other than the column names. But if we name the groups we get named columns out
pattern="(?P<First>^[\w]*)(?:.* )(?P<Last>[\w]*$)"

# Now call extract
names=df["President"].str.extract(pattern).head()
names
```

```python
# And we can just copy these into our main dataframe if we want to
df["First"]=names["First"]
df["Last"]=names["Last"]
df.head()
```
