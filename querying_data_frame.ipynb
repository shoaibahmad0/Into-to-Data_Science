```python
#importing pandas library

import pandas as pd

#bringing csv file

df =pd.read_csv('datasets/Admission_Predict.csv', index_col =0)

#removing white spaces from the strngs

df.columns = [x.lower().strip() for x in df.columns]

# now looking the result
df.head()
```

```python
admit_mask=df['chance of admit'] > 0.7
admit_mask
```

```python
#we only keep the data where it is true using .where() function

df.where(admit_mask)
```

```python
#now remove the nan data using dropna() function

df.where(admit_mask).dropna()
```

```python
# pandas use short syntax for combining the where() and dropna() function

df[df['chance of admit'] > 0.7]
```

```python
# applying index operator to get columns from data frame
df['chance of admit']
```

```python
#we can send it columns names as strings
df[['gre score','cgpa']]
```

```python
#now using boolean mask

df[df['gre score'] > 320]
```

```python
# combing two boolean mask 
(df['chance of admit'] > 0.7) & (df['chance of admit'] < 0.9)
```
