```python
# merging data

import pandas as pd
#staff data frame
staff_df = pd.DataFrame([{'name': 'kelly', 'Role':'Directorate of HR'},
                         {'name': 'Sally', 'Role':'Course liasian'},
                         {'name': 'James', 'Role':'Grader'}])

#set index by name

staff_df = staff_df.set_index('name')

#student data frame

stud_df=pd.DataFrame([{'name': 'James', 'School':'Buisness'},
                         {'name': 'Mike', 'School':'Law'},
                         {'name': 'Sally', 'School':'Engenering'}]) 

#now set index by name too

stud_df=stud_df.set_index('name')

print(staff_df)
print(stud_df)
```

```python
#now we can get union of these data frame by using merger() setting the how parameter to outer
pd.merge(staff_df, stud_df, how='outer', right_index=True,left_index=True)
```

```python
#now we can get intersection of these data frame by using merger()  set the how parameter of inner
pd.merge(staff_df, stud_df, how='inner', right_index=True,left_index=True)
```

```python
#getting the student regardless of whether they're are staff or not . just set how parameter to left
pd.merge(staff_df, stud_df, how='left', right_index=True,left_index=True)
```

```python
#getting the student if they are staff jsut set how parameter to true
pd.merge(staff_df, stud_df, how='right', right_index=True,left_index=True)
```

```python
#now first reset both dataframe

staff_df=staff_df.reset_index()
stud_df=stud_df.reset_index()


# now use on parameter instead of columns 

pd.merge(staff_df,stud_df, how='right', on='name')
```
